### Which Airport has the highest Origin Flights?

```python
plt.figure(figsize=(15,10))
sns.countplot('OriginAirportName',data=data_Set,order= data_Set['OriginAirportName'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

### To which Airport Arrival traffic is more

```python
plt.figure(figsize=(15,10))
sns.countplot('DestAirportName',data=data_Set,order= data_Set['DestAirportName'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

#### Which month has the highest traffic

```python
data_Set['Year'].value_counts()
```

```python
temp_dec = {1:'January',2:'Feb',3:'March',4:'April',5:'May',6:'June',7:'July',8:'Aug',9:'Sept',10:"Oct",11:'Nov',12:'Dec'}
```

```python
data_Set['Month'] = data_Set['Month'].apply(lambda x : temp_dec[x])
```

```python
sns.countplot(data_Set['Month'],order=data_Set['Month'].value_counts().index)
```

#### Which Airport is busiest

```python
city = data_Set.groupby(['OriginCity','Month'])['OriginAirportID'].count()
```

```python
city = city.to_frame()
```

```python
city.reset_index(inplace=True)
```

```python
city=city.rename({'OriginAirportID':'Total Number of Flights'},axis=1)
```

```python
city.sort_values('Total Number of Flights',ascending=False,inplace=True)
```

```python
city.head(20)
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(x='OriginCity',y='Total Number of Flights',data=city)
plt.xticks(rotation=90)
plt.show()
```

#### Cancelled Flight

```python
cancelled_flights = data_Set[data_Set['Cancelled']== 1]
```

```python
cancelled_flights['OriginAirportName'].value_counts()
```

```python
cancelled_flights['Carrier'].value_counts()
```

```python
Airport = cancelled_flights.loc[cancelled_flights['Cancelled'] == 1 , ['Carrier','OriginAirportName','OriginCity','DestAirportName','DestCity']]
```

```python
carrier_analysis = Airport.groupby(['Carrier','OriginAirportName','DestAirportName'])['DestCity'].count()
```

```python
carrier_analysis = carrier_analysis.to_frame()
```

```python
carrier_analysis.reset_index(inplace=True)
```

```python
carrier_analysis.head()
```

```python
carrier_analysis.rename({'DestCity':'No.Time Cancelled'},axis=1,inplace=True)
```

```python
carrier_analysis = carrier_analysis.sort_values('No.Time Cancelled',ascending=False)
```

```python
carrier_analysis.head()
```

```python
# Set CSS properties for th elements in dataframe
th_props = [
  ('font-size', '11px'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', '#6d6d6d'),
  ('background-color', '#f7f7f9')
  ]

# Set CSS properties for td elements in dataframe
td_props = [
  ('font-size', '11px')
  ]

# Set table styles
styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]
```

```python
cm = sns.light_palette("Red", as_cmap=True)
(carrier_analysis.style
  .background_gradient(cmap=cm, subset=['No.Time Cancelled'])
  .highlight_max(subset=['No.Time Cancelled'])
  .set_table_styles(styles))
```
